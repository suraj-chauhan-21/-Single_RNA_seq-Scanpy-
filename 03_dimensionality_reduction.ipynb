# 03 — Dimensionality Reduction

This notebook covers:
- PCA and elbow/scree plot
- k-nearest neighbour (kNN) graph
- UMAP embedding
- t-SNE embedding


```python
import sys, os
sys.path.insert(0, os.path.abspath('..'))

import scanpy as sc
import matplotlib.pyplot as plt
import config as cfg
from utils import load_adata, save_adata, ensure_dirs
from utils import plot_pca_variance, plot_umap, plot_tsne

sc.settings.verbosity = 2
sc.settings.set_figure_params(dpi=100, facecolor='white')
%matplotlib inline
print('Setup complete ✓')
```

```python
adata = load_adata(cfg.ADATA_NORMALIZED_PATH)
dr = cfg.DIM_REDUCTION
print(adata)
```

## 1. PCA

```python
sc.tl.pca(adata, n_comps=dr['n_comps_pca'], svd_solver='arpack', random_state=dr['random_state'])
print(f'PCA computed: {dr["n_comps_pca"]} components')
```

```python
# Scree plot — choose n_pcs where the curve flattens
plot_pca_variance(adata, n_pcs=50)
```

```python
sc.pl.pca(adata, color=['total_counts', 'pct_counts_mt'], ncols=2)
```

## 2. Neighbourhood Graph

Adjust `n_pcs` in `config.py` based on the scree plot above.

```python
sc.pp.neighbors(
    adata,
    n_neighbors=dr['n_neighbors'],
    n_pcs=dr['n_pcs'],
    random_state=dr['random_state'],
)
print(f'kNN graph built (k={dr["n_neighbors"]}, n_pcs={dr["n_pcs"]})')
```

## 3. UMAP

```python
sc.tl.umap(adata, min_dist=dr['umap_min_dist'], spread=dr['umap_spread'], random_state=dr['random_state'])
sc.pl.umap(adata, color=['total_counts', 'pct_counts_mt', 'n_genes_by_counts'], ncols=3)
```

## 4. t-SNE

```python
sc.tl.tsne(adata, n_pcs=dr['n_pcs'], perplexity=dr['tsne_perplexity'], random_state=dr['random_state'])
sc.pl.tsne(adata, color=['total_counts', 'pct_counts_mt'], ncols=2)
```

## 5. Save

```python
save_adata(adata, cfg.ADATA_REDUCED_PATH)
print('Done ✓')
```
